```python
api_url = 'https://coinmarketcap.com/currencies/{name_lower}/historical-data/?start=20100101&end=20180707'
for i,row in enumerate(master.to_dict(orient='records')):
    name = row['name']
    # Skip if we already have pricing data
#     if db['sp_price_history'].find_one({'name': name, 'source': 'CoinMarketCap'}):
#         print('Found existing pricing data for {name}. Skipping...'.format(name=name))
#         continue
        
    # Transform name (replace whitespace, lowercase)
    name_lower = name.replace(' ','-').lower()
    
    try:
        # Get price data from CoinMarketCap
        print('Getting Price Data from CoinMarketCap for {name} - {name_lower} ({i} of {length})'.format \
              (name=name,name_lower=name_lower,i=i+1,length=len(data)))
        response = requests.get(api_url.format(name_lower=name_lower))
        soup = BeautifulSoup(response.text,'html5lib')

        symbol = [x.text for x in soup.find_all("span", class_="text-bold h3 text-gray text-large")][0][1:-1]
        print('Writing records to db...')
        db.sp_price_history.delete_many({'symbol': symbol, 'source': 'CoinMarketCap'})

        tags = [x.text for x in soup.find_all("td")]
        cols = ['date','open','high','low','close','volume','cap']
        i = 0
        d = {
            'name': name,
            'symbol': symbol,
            'source': 'CoinMarketCap'
        }
        for i,t in enumerate(tags):
            colname = cols[i%len(cols)]
            if colname == 'date':
                d['close_date'] = datetime.strptime(t,'%b %d, %Y').strftime('%Y-%m-%d')
                d['price'] = None
            if colname == 'close':
                d['price'] = float(t)
                if d['price'] > 0:
                    db.sp_price_history.insert_one(d)
    except:
        # Fail silently
        pass
   
    time.sleep(2)
```
